```python
double = lambda x : x *2
print(double(6))
```

```python
add = lambda x,y : x + y
print(add(5,6))
```

```python
max_value = lambda x,y : x if x>y else y
print(max_value(5,6))
min_value = lambda x,y : x if x < y else y
print(min_value(5,6))
```

```python
fullname = lambda first, last: first + " " + last
print(fullname("James", "Varadha"))
```

```python
is_even = lambda E : E % 2 == 0
print(is_even(5))
print(is_even(6))
```

```python
age_check = lambda age : True if age >= 18 else False
print(age_check(17))
print(age_check(18))
```

```python
#map(function,collection)

def c_to_f(temp):
 return (temp * 9/5) + 32

celcius_temps =[0.0, 10.0,20.0,30.0,40.0,50.0]
print(celcius_temps)
farenhiet_temps = list(map(c_to_f, celcius_temps))
print(farenhiet_temps)
# or use lambda instead of Def
farenhiet_temps = list(map(lambda temp : (temp * 9/5) + 32, celcius_temps))
print(farenhiet_temps)
```

```python
import pandas as pd
data =[100,102,103,300,400,500]
series = pd.Series(data,index=["a","b","c","d","e","f"])
print(series)
series.loc["c"] = 200
print(series)
print(series.iloc[0]) # For integer location
print(series.iloc[1:4:2])
print(series.iloc[-1])
print(series[series>200])
print(series[series<200])
```

```python
calories={"day1":1300,"day2":1550,"day3":1700}
myvar=pd.Series(calories)
print(myvar)
print(myvar["day1"])
print(myvar.loc["day1"])
print(myvar.iloc[0])
myvar.loc["day3"]+= 200
print(myvar)
```
